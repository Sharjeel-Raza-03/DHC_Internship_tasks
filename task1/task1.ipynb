## Task 1: Exploring and Visualizing the Iris Dataset

## Objective
The objective of this task is to load, inspect, summarize, and visualize the Iris dataset to understand data trends and distributions.

## Dataset
The Iris dataset contains measurements of iris flowers including sepal length, sepal width, petal length, petal width, and species.

## Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Better plot style
sns.set(style="whitegrid")
```

## Step 2: Load Dataset

```python
df = sns.load_dataset("iris")
```

## Step 3: Inspect Dataset

```python
print("Dataset shape:", df.shape)
```

## Step 5: Data Visualization

```python
df["species"].value_counts()
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(
    data=df,
    x="sepal_length",
    y="sepal_width",
    hue="species",
    style="species",
    s=80
)

plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(
    data=df,
    x="petal_length",
    y="petal_width",
    hue="species",
    style="species",
    s=80
)

plt.title("Petal Length vs Petal Width")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```

```python
sns.pairplot(df, hue="species", diag_kind="hist")
plt.show()
```

```python
df.hist(figsize=(10, 8), bins=20, color="skyblue", edgecolor="black")

plt.suptitle("Feature Distributions in Iris Dataset", fontsize=16)
plt.show()
```

```python
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

for feature in features:
    plt.figure(figsize=(8, 5))
    
    sns.histplot(
        data=df,
        x=feature,
        hue="species",
        kde=True,
        bins=20
    )
    
    plt.title(f"Distribution of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.boxplot(data=df.drop("species", axis=1))

plt.title("Box Plot of Iris Features")
plt.xlabel("Features")
plt.ylabel("Values")
plt.show()
```

```python
for feature in features:
    plt.figure(figsize=(8, 5))
    
    sns.boxplot(
        data=df,
        x="species",
        y=feature
    )
    
    plt.title(f"{feature} by Species")
    plt.xlabel("Species")
    plt.ylabel(feature)
    plt.show()
```

```python
numeric_df = df.drop("species", axis=1)

correlation = numeric_df.corr()

plt.figure(figsize=(8, 6))

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap of Iris Features")
plt.show()
```

## Step 6: Final Insights

- The Iris dataset contains 150 rows and 5 columns.
- There are 4 numerical features: sepal_length, sepal_width, petal_length, and petal_width.
- The target/category column is species.
- The dataset has three species: setosa, versicolor, and virginica.
- Petal length and petal width show clear separation between species.
- Setosa is the easiest species to separate visually.
- Box plots show the spread of each feature and help identify possible outliers.
- Histograms show that feature values differ across species.
